```python
// Perguntas para cada nível
const perguntas = {
    facil: [
        {
            pergunta: "Qual é o símbolo químico do oxigênio?",
            resposta: "O"
        },
        {
            pergunta: "Quantos elementos químicos existem na tabela periódica?",
            resposta: "118"
        },
        // Adicione mais perguntas fáceis
    ],
    medio: [
        {
            pergunta: "O que é um ácido?",
            resposta: "Substância que libera íons de hidrogênio em solução aquosa"
        },
        {
            pergunta: "Quais são os três estados físicos da matéria?",
            resposta: "Sólido, líquido e gasoso"
        },
        // Adicione mais perguntas médias
    ],
    dificil: [
        {
            pergunta: "O que são isótopos?",
            resposta: "Átomos de um mesmo elemento com números diferentes de nêutrons"
        },
        {
            pergunta: "Quais são as leis ponderais da química?",
            resposta: "Lei de conservação da massa, Lei das proporções definidas e Lei das proporções múltiplas"
        },
        // Adicione mais perguntas difíceis
    ]
};

// Função para apresentar e verificar respostas
function jogarNivel(perguntasNivel) {
    let pontos = 0;

    for (let i = 0; i < perguntasNivel.length; i++) {
        const respostaUsuario = prompt(perguntasNivel[i].pergunta);

        if (respostaUsuario.toLowerCase() === perguntasNivel[i].resposta.toLowerCase()) {
            alert("Resposta correta!");
            pontos++;
        } else {
            alert("Resposta incorreta!");
        }
    }

    alert(`Fim do nível. Pontuação: ${pontos}/${perguntasNivel.length}`);
}

// Início do jogo
alert("Bem-vindo ao jogo de perguntas e respostas de Química!");

const nivelEscolhido = prompt("Escolha o nível de dificuldade: Fácil, Médio ou Difícil").toLowerCase();

switch (nivelEscolhido) {
    case "fácil":
        jogarNivel(perguntas.facil);
        break;
    case "médio":
        jogarNivel(perguntas.medio);
        break;
    case "difícil":
        jogarNivel(perguntas.dificil);
        break;
    default:
        alert("Nível inválido. Recarregue a página para tentar novamente.");
}
```
